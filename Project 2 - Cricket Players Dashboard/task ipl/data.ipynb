```python
colsToInt = ['Centuries', 'Half_Centuries', 'Fours', 'Sixes', 'Catches_Taken', 'Stumpings',
        'Matches_Bowled', 'Balls_Bowled', 'Runs_Conceded', 'Wickets_Taken',
        'Four_Wicket_Hauls', 'Five_Wicket_Hauls']

for col in colsToInt:
    cricketDf[col] = pd.to_numeric(cricketDf[col], errors="coerce").astype("Int64")
    cricketDf[col] = cricketDf[col].fillna(0)
```

```python
colsToFloat = ['Bowling_Average', 'Economy_Rate', 'Bowling_Strike_Rate']

for col in colsToFloat:
    cricketDf[col] = cricketDf[col].replace("No stats", "0")
    cricketDf[col] = cricketDf[col].astype(float).fillna(0)
```
